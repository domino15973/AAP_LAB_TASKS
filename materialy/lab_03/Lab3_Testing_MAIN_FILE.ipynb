## 1. Wstęp: Dlaczego testujemy?

Wyobraź sobie, że budujesz dom. Czy wprowadziłbyś się do niego bez inspekcji budowlanej? Pewnie nie — chciałbyś mieć pewność, że ściany są proste, instalacja elektryczna bezpieczna, a dach nie przecieka.

**Testy w programowaniu pełnią dokładnie tę samą rolę** — sprawdzają, czy Twój kod działa zgodnie z oczekiwaniami, zanim trafi na produkcję.

**Trzy kluczowe korzyści z testowania:**
1. **Wczesne wykrywanie błędów** — im wcześniej znajdziesz błąd, tym taniej i szybciej go naprawisz.
2. **Dokumentowanie zachowania** — test jest żywą dokumentacją, która mówi: *"ta funkcja powinna zwracać X dla danych Y"*.
3. **Bezpieczne zmiany** — refaktoryzujesz kod? Testy powiedzą Ci, czy niczego nie zepsułeś.

W Pythonie mamy dwa główne frameworki testowe:
- **unittest** — wbudowany w Pythona (nie trzeba nic instalować)
- **pytest** — zewnętrzna biblioteka (prostszy, bardziej nowoczesny)

Na dzisiejszych zajęciach poznamy oba i porównamy je na praktycznych przykładach.

## 2. Słowo kluczowe `assert`

Zanim przejdziemy do frameworków testowych, poznajmy podstawę: słowo kluczowe **`assert`**.

`assert` sprawdza, czy podany warunek jest prawdziwy. Jeśli nie — rzuca wyjątek `AssertionError`.

```python
assert warunek, "Opcjonalna wiadomość o błędzie"
```

To fundament, na którym opiera się całe testowanie w Pythonie.

````python
# Basic usage of assert (English version)
x = 5
assert x == 5, "x should be 5"
print("Assertion passed: x == 5")

# What happens when the condition is NOT met?
try:
    assert x == 10, "x is not 10!"
except AssertionError as e:
    print(f"Assertion error caught: {e}")

# Practical example -- testing a simple function
def add(a, b):
    return a + b

assert add(2, 3) == 5
assert add(-1, 1) == 0
assert add(0, 0) == 0
print("All assertions passed successfully!")
````

## 3. unittest — wbudowany framework testowy

**unittest** to moduł wbudowany w Pythona (nie trzeba nic instalować). Jest wzorowany na frameworku JUnit z Javy.

**Kluczowe zasady unittest:**
- Testy organizujemy w **klasy** dziedziczące po `unittest.TestCase`
- Każda metoda testowa **musi zaczynać się od `test_`** (inaczej nie zostanie wykryta!)
- Używamy specjalnych metod asercji zamiast zwykłego `assert`

### Najpopularniejsze asercje unittest:

| Metoda | Sprawdza |
|--------|----------|
| `assertEqual(a, b)` | a == b |
| `assertNotEqual(a, b)` | a != b |
| `assertTrue(x)` | bool(x) is True |
| `assertFalse(x)` | bool(x) is False |
| `assertRaises(exc)` | czy rzucono wyjątek |
| `assertIn(a, b)` | a in b |

### Demo: Testowanie klasy Calculator z unittest

Poniżej znajdziesz klasę `Calculator` oraz jej testy jednostkowe.

**WAŻNE:** Testy unittest najlepiej uruchamiać z terminala (nie z Jupytera), ponieważ `unittest.main()` kończy proces. Dlatego kod klasy i testów znajduje się w osobnych plikach:
- `calculator.py` — klasa, którą testujemy
- `test_calculator_unittest.py` — testy

Poniżej pokazujemy zawartość obu plików, a następnie uruchamiamy testy komendą terminalową.

````python
# ========================================
# Zawartosc pliku: calculator.py
# ========================================

class Calculator:
    """Simple calculator with four basic operations."""

    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            raise ValueError("Division by zero!")
        return a / b

# Manual test
calc = Calculator()
print(f"2 + 3 = {calc.add(2, 3)}")
print(f"10 - 5 = {calc.subtract(10, 5)}")
print(f"3 * 7 = {calc.multiply(3, 7)}")
print(f"10 / 2 = {calc.divide(10, 2)}")
print("\nCalculator class working correctly!")
````

````python
!python -m unittest test_calculator_unittest -v
````
